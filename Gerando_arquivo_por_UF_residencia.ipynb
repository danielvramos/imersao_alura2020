```python
nome_arquivo = 'MICRODADOS_ENEM_2019.csv'
uf = 'RJ'
```

```python
import csv

def gerador(arq, uf):
    with open(arq, "r") as csvfile:
        datareader = csv.reader(csvfile)
        yield next(datareader)[0].split(";")      
        for row in datareader:
            if row[0].split(";")[5] == uf:
                yield row[0].split(";")
```

```python
lista = list(gerador(nome_arquivo,uf ))
```

```python
len(lista)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(lista)
```

```python
df.columns = df.loc[0]
df.drop(0, inplace =True )
```

```python
df
```

```python
df.to_csv(uf+"_"+nome_arquivo,sep=";",index=False)
```
